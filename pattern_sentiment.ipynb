```python
from pattern.en import sentiment
import pandas
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

original_ratings = pandas.read_csv('id-rating.csv', sep=",")
p_comments = pandas.read_csv('P_1917.csv', sep=",")
```

```python
movie_id = p_comments['Video ID'][1]
movie_row = original_ratings.loc[original_ratings['ID'] == movie_id]
movie_rating = movie_row['Rating']
```

```python
def find_pol(review):
    return sentiment(review)[0]

p_comments['Sentiment_Polarity'] = p_comments['Comment'].apply(find_pol)
p_comments = p_comments[p_comments['Sentiment_Polarity'] !=0]
plot = sns.distplot(p_comments['Sentiment_Polarity'])

y = plot.lines[0].get_ydata()
x = plot.lines[0].get_xdata()
maxid = np.argmax(y)
print('Movie: 1917')
print('IMDB rating: ',  movie_rating.values[0], 'Sentiment: ', x[maxid])
```

```python
bad = 0
good = 0
neutral = 0

for index, row in p_comments.iterrows():
    comment = row['Comment']
    s_comment = sentiment(comment)
    print(s_comment)
    if s_comment[0] >= 0.4:
        good += 1
    elif s_comment[0] < 0.4 and s_comment[0] > -0.25:
        neutral += 1
    else:
        bad +=1

objects = ('Good', 'Neutral', 'Bad')
y_pos = np.arange(len(objects))
performance = [good, neutral, bad]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Number of comments')
plt.title('Distribution of comment sentiment')

plt.show()
```
